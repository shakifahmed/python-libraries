```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('default')
```

```python
df = sns.load_dataset('tips') # 'tips' is built in dataset
df.head()
```

# relplot() function

```python
sns.relplot(x='total_bill',y='tip',data=df)
```

# hue parameter
The hue parameter in seaborn is used to specify a categorical variable to apply different colors for the different levels of the variable.

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=df)
```

# style parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=df)
```

# catagorical plot
catplot() function

x axis = Catagorical plot & 
y axis = numerical plot

```python
# strip plot
sns.catplot(x='day',y='tip',kind='strip',data=df)
```

```python
# swram plot
# sns.catplot(x='day',y='tip',kind='swarm',data=df)

# canbe write this way
sns.swarmplot(x='day',y='tip',data=df)
```

```python
sns.stripplot(x='day',y='tip',hue='smoker',data=df)
```

# boxplot() function

```python
sns.boxplot(df['tip'])
```

```python
sns.boxplot(x='day',y='tip',data=df)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=df)
```

# violinplot() function

```python
sns.violinplot(x='day',y='total_bill',hue='sex',split=True,data=df)
```

# heatmap() function

```python
#first need to make a pivot table
hm = df.pivot_table(index='day',columns='sex',values='tip',aggfunc='sum')
hm
```

```python
sns.heatmap(hm,annot=True,fmt='.2f')
```

# jointplot() function

```python
sns.jointplot(x='total_bill',y='tip',data=df) # kind = 'scatter'
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='reg',data=df)
```
